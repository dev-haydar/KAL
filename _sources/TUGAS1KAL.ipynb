## Pengertian Sistem Persamaan Linier
Sistem Persamaan Linier (SPL) adalah kumpulan dari dua atau lebih persamaan linier yang memiliki variabel yang sama dan harus diselesaikan secara bersamaan. Persamaan linier sendiri merupakan persamaan matematika di mana setiap variabel berpangkat satu dan tidak terdapat perkalian antar variabel.

Secara umum, SPL dengan dua variabel dapat dituliskan sebagai:

$$
\begin{cases} 
a_1x + b_1y = c_1 \\ 
a_2x + b_2y = c_2 
\end{cases}
$$

## Mengapa Sistem Persamaan Linier Penting?

SPL memiliki banyak aplikasi dalam kehidupan sehari-hari, mulai dari bidang teknik, ekonomi, hingga ilmu komputer. Contohnya:

* Di bidang ekonomi, SPL digunakan untuk menentukan keseimbangan antara permintaan dan penawaran.
* Dalam teknik dan fisika, SPL digunakan untuk menganalisis rangkaian listrik dan sistem mekanik.
* Di ilmu komputer, SPL sering diterapkan dalam kecerdasan buatan dan pemrosesan data.

## Solusi Persamaan Linier

1. Penyelesaian Tunggal (Unique Solution)
Definisi: Sistem persamaan linier memiliki satu solusi jika garis-garis yang merepresentasikan persamaan tersebut berpotongan di satu titik.

Contoh: Misalkan kita memiliki sistem persamaan berikut: 

$$
\begin{aligned}
2x + 3y &= 6 \quad &(1) \\
x - y &= 1 \quad &(2)
\end{aligned}
$$

Jika kita menyelesaikan sistem ini, kita akan menemukan satu titik \((x, y)\) yang memenuhi kedua persamaan.  
Misalnya, solusi dari sistem ini adalah: 

$
x = 3, \quad y = 0
$

2. Banyak Penyelesaian (Infinite Solutions)
Definisi: Sistem persamaan linier memiliki banyak solusi jika garis-garis yang merepresentasikan persamaan tersebut tumpang tindih (berimpit).

Contoh: Misalkan kita memiliki sistem persamaan berikut: 

$$
\begin{align*}
2x + 4y &= 8 \quad (1) \\
x + 2y &= 4 \quad (2)
\end{align*}
$$

Persamaan (1) dapat disederhanakan menjadi \( x + 2y = 4 \), yang sama dengan persamaan (2).  
Karena kedua persamaan ini merepresentasikan garis yang sama, maka ada **banyak solusi**, yaitu semua pasangan \((x, y)\) yang memenuhi: 

$
x + 2y = 4
$

3. Tidak Ada Penyelesaian (No Solution)
Definisi: Sistem persamaan linier tidak memiliki solusi jika garis-garis yang merepresentasikan persamaan tersebut sejajar dan tidak pernah berpotongan.

Contoh: Misalkan kita memiliki sistem persamaan berikut: 

$$
\begin{align*}
2x + 3y &= 6 \quad (1) \\
2x + 3y &= 8 \quad (2)
\end{align*}
$$

Kedua persamaan ini memiliki koefisien yang sama untuk \( x \) dan \( y \), tetapi konstanta di sisi kanan berbeda.  
Ini berarti kedua garis **sejajar** dan **tidak akan pernah berpotongan**, sehingga **tidak ada solusi** untuk sistem ini.

### 1. Metode Substitusi
**Definisi:** Metode substitusi adalah teknik menyelesaikan sistem persamaan dengan menyatakan salah satu variabel dalam bentuk variabel lain dari satu persamaan, lalu memasukkannya ke persamaan lainnya.

**Contoh Sistem Persamaan:**
$$
\begin{aligned}
2x + 3y &= 6 \quad &(1) \\
x - y &= 1 \quad &(2)
\end{aligned}
$$

**Langkah-langkah Penyelesaian:**

1. **Ubah persamaan (2)** untuk menyatakan $x$ dalam bentuk $y$:
   $$x = y + 1$$

2. **Substitusi** nilai $x$ tersebut ke dalam **persamaan (1)**:
   $$
   \begin{aligned}
   2(y + 1) + 3y &= 6 \\
   2y + 2 + 3y &= 6 \\
   5y + 2 &= 6 \\
   5y &= 4 \\
   y &= \frac{4}{5}
   \end{aligned}
   $$

3. **Substitusi kembali** nilai $y = \frac{4}{5}$ ke dalam persamaan $x = y + 1$:
   $$
   \begin{aligned}
   x &= \frac{4}{5} + 1 \\
   x &= \frac{4}{5} + \frac{5}{5} \\
   x &= \frac{9}{5}
   \end{aligned}
   $$

**Himpunan Penyelesaian:** $x = \frac{9}{5}, y = \frac{4}{5}$

**Solusi:**

$
\left( x = \frac{9}{5}, \quad y = \frac{4}{5} \right)
$


### 2. Eliminasi Gauss
**Definisi:** Eliminasi Gauss adalah metode sistematis untuk menyelesaikan sistem persamaan linier dengan mengubah matriks koefisien menjadi bentuk **eselon baris** (segitiga atas) menggunakan Operasi Baris Elementer (OBE).

**Contoh:** Misalkan kita memiliki sistem persamaan berikut:
$$
\begin{aligned}
2x + 3y + z &= 1 \quad &(1) \\
4x + y - z &= 2 \quad &(2) \\
-2x + y + 2z &= 3 \quad &(3)
\end{aligned}
$$

**Langkah-langkah:**

1. **Representasi Matriks Augmented:**
   $$
   \left[
   \begin{array}{ccc|c}
   2 & 3 & 1 & 1 \\
   4 & 1 & -1 & 2 \\
   -2 & 1 & 2 & 3
   \end{array}
   \right]
   $$

2. **Operasi Baris Elementer (OBE):**
   Setelah melakukan serangkaian operasi baris (seperti $R_2 - 2R_1$ dan $R_3 + R_1$), matriks akan berubah menjadi bentuk segitiga atas:
   $$
   \left[
   \begin{array}{ccc|c}
   2 & 3 & 1 & 1 \\
   0 & -5 & -3 & 0 \\
   0 & 0 & 1 & 1
   \end{array}
   \right]
   $$

3. **Substitusi Mundur:**
   Dari baris terakhir, kita mendapatkan $z = 1$. Masukkan nilai $z$ ke baris kedua untuk mencari $y$, lalu masukkan $y$ dan $z$ ke baris pertama untuk mencari $x$.

### 3. Solusi Grafik
**Definisi:** Metode grafik dilakukan dengan menggambar setiap persamaan linier pada koordinat Kartesius. Titik di mana garis-garis tersebut saling berpotongan merupakan solusi dari sistem tersebut.

**Contoh Sistem Persamaan:**
$$
\begin{aligned}
y &= 2x + 1 \quad &(1) \\
y &= -x + 4 \quad &(2)
\end{aligned}
$$

**Interpretasi Hasil:**
* **Satu Titik Potong:** Sistem memiliki solusi tunggal (konsisten).
* **Garis Sejajar:** Sistem tidak memiliki solusi (inkonsisten).
* **Garis Berhimpit:** Sistem memiliki solusi tak hingga.
